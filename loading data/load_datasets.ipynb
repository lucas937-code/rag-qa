# Split the dataset

```python
from datasets import load_dataset

# Load the original dataset
trivia_qa = load_dataset("trivia_qa", name="rc.wikipedia")

# The HF dataset has:
#   trivia_qa["train"]
#   trivia_qa["validation"]  --> this becomes test

# Take first 7900 examples for validation
val_dataset = trivia_qa["train"].select(range(7900))

# Remaining examples form the new training set
train_dataset = trivia_qa["train"].select(range(7900, len(trivia_qa["train"])))

# Test set is the original validation set
test_dataset = trivia_qa["validation"]

print("Original train:", len(trivia_qa["train"]))
print("New train:", len(train_dataset))
print("Validation:", len(val_dataset))
print("Test:", len(test_dataset))
```

```python
train_dataset.save_to_disk("../train_dataset/")
val_dataset.save_to_disk("../validation_dataset/")
test_dataset.save_to_disk("../test_dataset/")
```
